```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('../Datasets/EDOS 1M.csv')
X,y = dataset["uttr"], dataset["eb+_emot"]

classes = y.unique()
counts = pd.get_dummies(y).sum()

plt.ylabel('Counts')
plt.xlabel('Classes')
sns.barplot(x = [i for i in range(len(classes))], y = counts)
plt.show()
```

```python
num_datasets = 4
min_samples = counts.min()

# building num_datasets different datasets taking the same number of samples for each class
# we'll take the minimum number of samples -> minority class will be completely included in all datasets
for i in range(num_datasets):
    tmp_df = pd.DataFrame()
    for label in classes:
        tmp_df = pd.concat([tmp_df, dataset[dataset["eb+_emot"] == label].sample(min_samples)])

    print("Dataset " + str(i) + " created")
    tmp_df.to_csv('../Datasets/EDOS 1M - ' + str(i) + '.csv', index=False)
```
